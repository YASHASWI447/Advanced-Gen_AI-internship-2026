# Problem 1: Social Media – Post Engagement Analyzer


```python
# create function to analyze post engagemnet
def analyze_post(likes):
    total = 0
    for i in likes:
        total = total + i
    # check post status
    if total >= 1000:
        status = "Viral"
    else:
            status = "Normal"
    return total,status
# user input
n = int(input("Enter the number of likes:"))
likes = []

for i in range(n):
    num = int(input("Enter like : "))
    likes.append(num)
# call function
result = analyze_post(likes)
# Display result
print("Total Likes:",result[0])
print("Post Status:",result[1])
```

# Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def check_rainfall(rain_list, required):
    total = 0
    for rain in rain_list:
        total = total + rain
    average = total / len(rain_list)
    if average >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    return average,status

# user input
n = int(input("Enter number of readings:"))
rain_list = []

for i in range(n):
    value = int(input("Enter rainfall:"))
    rain_list.append(value)
required = int(input("Enter required rainfall:"))
# function call
avg, result = check_rainfall(rain_list, required)

print("Average Rainfall:",avg)
print("Rainfall  Status:",result)

    
```

# Problem 4: Social Media – Duplicate Account Detection


```python
# function to check duplicate names
def check_duplicates(names):
    unique_names  = set(names)
    if len(names) != len(unique_names):
        print("Duplicate Accounts Found: Yes")
    else:
        print("All accounts are Unique")
# Take usernames from user
user_input = input("Enter usernames seperated by space:")
# convert input into list
names_list = user_input.split()
# call function
check_duplicates(names_list)    
```

# Problem 6: Agriculture – Premium Crop Price Filter


```python
# function to find premium crops
def premium_crops(prices):
    premium = [] # empty list for premium crops

    for price in prices:
        if price > 2000:
            premium. append(price)
    print("Premium Crops:",premium)

# Take user input
prices = list(map(int,input("Enter  crop prices(space separated): ").split()))
# call function
premium_crops(prices)
            

```

# Problem 7: System Monitoring – Application Health Checker

```python
def health_checker(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor issues"
    else:
        status  = "Critical Issues"
    print("Error Count:",errors)
    print("System Status:",status)
# take user input
errors = int(input("Enter number of errors:"))
# call function
health_checker(errors)
```

# Problem 9: E-Learning – Student Attendance Eligibility System


```python
# Function to check attendance
def attendance_eligibility(total,attended):
    percentage = (attended/total) * 100
    # condition to check eligibility
    if percentage >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"
    print("Attendance Percentage:",percentage)
    print("Exam Eligibility:",result)
# User input
total_classes =  int(input("Enter total classes:"))
attended_classes = int(input("Enter attended classes:"))
# function call
attendance_eligibility(total_classes,attended_classes)
    
```

# Problem 10: Smart Electricity Bill Analyzer


```python
# Function to calculate electricity bill
def calculate_bill(units):

    bill = 0

    # First 100 units
    if units <= 100:
        bill = units * 3

    # Next 100 units
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    # Above 200 units
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        status = "Low Usage"

    elif bill <= 1500:
        status = "Moderate Usage"

    else:
        status = "High Usage"

    return bill, status


# Taking input from the user
units = int(input("Enter units consumed: "))

# Calling function
total_bill, usage_status = calculate_bill(units)

# Printing result
print("Total Bill:", total_bill)
print("Usage Status:", usage_status)
```
