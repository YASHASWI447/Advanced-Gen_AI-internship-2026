```python

# Problem 1: Mobile Recharge Validation System


def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return "Recharge amount must be at least ₹50."

    if plan not in valid_plans:
        return "Invalid plan selected."

    return "Recharge Successful!"

# Retry mechanism
while True:
    amount = int(input("Enter recharge amount: "))
    plan = int(input("Enter plan amount: "))

    result = validate_recharge(amount, plan)
    print(result)

    if result == "Recharge Successful!":
        break
    else:
        print("Please try again.\n")
```

```python

# Problem 2: Inventory Reorder Alert System


def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product} → Reorder Alert")
        else:
            print(f"{product} → Stock OK")

inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

check_inventory(inventory)
```

```python

# Problem 3: Student Result Processing System


def process_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average: {average:.2f} → Pass"
    else:
        return f"Average: {average:.2f} → Fail"

marks_list = [60, 45, 70, 50]
print(process_result(marks_list))
```

```python

# Problem 4: Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra

    return fare

# Retry system
while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)
    print(f"Total Fare: ₹{total_fare:.2f}")

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python

# Problem 5: Attendance Eligibility Checker


def check_eligibility(attendance):
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% → Eligible"
    else:
        return f"Attendance: {percentage:.2f}% → Not Eligible"

attendance_list = ["P", "P", "A", "P", "P"]
print(check_eligibility(attendance_list))
```

```python

# Problem 6: Password Strength Checker


def check_password(password):
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include at least one digit and one special character)"

user_password = input("Enter password: ")
print(check_password(user_password))
```
