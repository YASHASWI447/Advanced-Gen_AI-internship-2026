```python
#Problem 1
def smart_parking_system(capacity, vehicle_logs):
    current_parked = 0
    peak_usage = 0
    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
            if current_parked > peak_usage:
                peak_usage = current_parked
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1 
    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage) 
    if current_parked > capacity:
        print("Parking Status: Full - Capacity Exceeded!")
    elif current_parked == capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking_system(capacity, vehicle_logs)
```

```python
#Problem 2
def estimate_delivery_time(distance, traffic, weather):
    # Base speed (km per hour)
    base_speed = 20  
    
    # Base time calculation
    base_time = (distance / base_speed) * 60
    
    # Traffic delay
    if traffic.lower() == "low":
        traffic_delay = 0
    elif traffic.lower() == "medium":
        traffic_delay = 10
    elif traffic.lower() == "high":
        traffic_delay = 20
    else:
        traffic_delay = 0
    
    # Weather delay
    if weather.lower() == "clear":
        weather_delay = 0
    elif weather.lower() == "rainy":
        weather_delay = 11
    elif weather.lower() == "stormy":
        weather_delay = 25
    else:
        weather_delay = 0
    
    # Total estimated time
    total_time = base_time + traffic_delay + weather_delay
    
    return round(total_time)

distance = 8
traffic = "High"
weather = "Rainy"

eta = estimate_delivery_time(distance, traffic, weather)
print(f"Estimated Delivery Time: {eta} minutes")
```

```python
#Problem 3
def analyze_seat_occupancy(total_seats, booked_seats):
    
    # Count booked seats (count of 1s)
    booked_count = sum(booked_seats)
    
    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100
    
    # Determine show status
    if occupancy == 100:
        status = "Housefull"
        suggestion = "Open additional show immediately."
    elif occupancy >= 75:
        status = "Almost Full"
        suggestion = "Consider opening additional show."
    else:
        status = "Seats Available"
        suggestion = "No need for additional show."
    
    return round(occupancy), status, suggestion
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

# Function Call
occupancy, status, suggestion = analyze_seat_occupancy(total_seats, booked_seats)

print(f"Occupancy: {occupancy}%")
print(f"Show Status: {status}")
print(f"Suggestion: {suggestion}")
```

```python
#Problem 4
def classify_server_load(cpu_readings):
    
    # Calculate total using loop
    total = 0
    for reading in cpu_readings:
        total += reading
    
    # Calculate average
    average = total / len(cpu_readings)
    
    # Classification
    if average < 50:
        status = "Normal"
    elif 50 <= average <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    return round(average), status

cpu_readings = [45, 60, 70, 85, 90]

# Function Call
average_load, server_status = classify_server_load(cpu_readings)

print(f"Average CPU Load: {average_load}%")
print(f"Server Status: {server_status}")
```

```python
#Problem 5
def monitor_resource_usage(resource_usage):
    
    overused = []
    threshold = 8  # hours
    
    # Check each resource
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)
    
    # Determine energy alert
    if overused:
        energy_alert = "Yes"
    else:
        energy_alert = "No"
    
    return overused, energy_alert
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4}

# Function Call
overused_resources, alert = monitor_resource_usage(resource_usage)

print("Overused Resources:", ", ".join(overused_resources))
print("Energy Alert:", alert)
```

```python
#Problem 6
def manage_registrations(capacity, registrations):
    confirmed = 0
    waitlisted = 0
    # Loop to validate each registration
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    # Determine registration status
    if registrations >= capacity:
        status = "Closed"
    else:
        status = "Open"
    return confirmed, waitlisted, status
capacity = 100
registrations = 105
# Function Call
confirmed_users, waitlisted_users, registration_status = manage_registrations(capacity, registrations)
print(f"Confirmed Registrations: {confirmed_users}")
print(f"Waitlisted Users: {waitlisted_users}")
print(f"Registration Status: {registration_status}")
```
